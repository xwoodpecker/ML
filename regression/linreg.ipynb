# Least Squares in Python
Example: Prediction of fuel efficiency of a car based on its power. Usage of linear regression by fitting the following model:  
<br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; 
$\hat{y} = w_o + w_1 \cdot x$  
<br>
Where x corresponds to the horsepower and $\hat{y}$ corresponds to the fuel efficiency.

## Step 1: Data Import
Data will be imported using the pandas library.

```python
import pandas as pd #short name for library

cars = pd.read_csv('/Users/WoodPecker/PycharmProjects/JupyterProject/auto-mpg.data.txt', header=None, sep='\s+')
print(cars)
```

## Step 2: Extract relevant data
Miles per gallon and Horsepower will be extracted from our data. The numpy library contains the necessary data structure.
*cars.iloc[:,0]* &nbsp; - &nbsp; ":" is a slice expression for all rows, "0" defines the extracted column of the data

```python
import numpy as np
mpg = cars.iloc[:,0].values # returns a numpy.ndarray all rows, column 0 (miles per gallon)
hp = cars.iloc[:,3].values # returns a numpy.ndarray all rows, column 3 (horse power)
```

## Step 3: Plot the data
Matplotlib is the library that does the plotting. 
A scatter plot with power on the x axis and fuel consumption on the y axis is executed.

```python
import matplotlib.pyplot as plt
plt.scatter(hp, mpg, color='blue', marker='x')
plt.xlabel('Power [hp]')
plt.ylabel('Fuel consumption [miles/gallon]')
plt.show()

```

## Step 4: Compute and plot the regression line
The data is converted in a way that Sklearn can handle it. Then a regression model is fitted to our data. 
x represents our data and y is the value that we want to predict. A straight line is defined by two parameters: 
$w_o$ is the axis intercept and $w_1$ is the slope of the curve. These values are determined by the linear regression. 
Now we can add the regression line to our plot.

```python
from sklearn import linear_model
x = [] # we need an list of lists here
for i in hp:
    x.append([i])
y = mpg
reg = linear_model.LinearRegression()
reg.fit(x,y)
reg.intercept_ # returns 39.93586102117046 as the optimal w0
reg.coef_[0] # returns -0.15784473335365357 as the optimal w1


plt.scatter(hp, mpg, color='blue', marker='x')
plt.xlabel('Power [hp]')
plt.ylabel('Fuel consumption [miles/gallon]')
plt.plot(x, reg.predict(x), color='red', linewidth=2)
plt.show()

```
